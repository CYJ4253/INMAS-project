```python
from html import unescape
import numpy as np
from urllib.request import urlopen,Request
import time
import pandas as pd
import os

def VA_extract(VA_html):
    Shows = np.array(VA_html.split('<div class="spaceit_pad">'))
    substrcheck = np.vectorize(lambda a,A: a in A,otypes=[bool])
    Shows = Shows[np.where(substrcheck('class="js-people-title">',Shows))]
    def gt(x):
        try:
            return x.replace('</a>','class="js-people-title">').split('class="js-people-title">')[1]
        except:
            return "bruh"
    
    get_title = np.vectorize(gt,otypes=[str])
    return np.unique(get_title(Shows))

substrcheck = np.vectorize(lambda a,A: a in A,otypes=[bool])

def character_shows(character_html):
    Shows = np.array(character_html.split('</a>'))
    Shows = np.vectorize(lambda a: a.split(">")[-1],otypes=[str])(Shows[np.where(substrcheck('<td valign="top" class="borderClass">',Shows))])
    return np.unique(Shows)

def apply_to(thing,folder,D,fun):
    try:
        with open(folder+"/"+thing,"r",encoding='utf-8') as f:
            D[thing.split(".")[0]] = fun(unescape(f.read()))
    except PermissionError:
        pass
```

```python
from IPython.display import clear_output

def get_anime(character):
    time.sleep(abs(np.random.normal(2.,.5)))
    character_html = unescape(urlopen(Request(character)).read().decode('utf-8'))
    try:
        return character_shows(character_html)
    except:
        return ''

character_names = []
corresponding_anime = []
N = 500
sector = np.random.choice(characters[64000:],N) # change range 
for i in range(N):
    clear_output(wait=True)
    character_name = sector[i].split('/')[-1].replace('_',' ')
    try:
        show = get_anime(sector[i])
    except Exception as e:
        print(character_name,e)
        continue
    print("{}/{}".format(i+1,N))
    character_names.append(character_name)
    corresponding_anime.append(show)
    print(character_names[-1],corresponding_anime[-1])

characters_db = pd.DataFrame({'Names':character_names,'Anime':corresponding_anime})
characters_db
```

```python
import datetime

months = {'Jan':1,'Feb':2,'Mar':3,'Apr':4,'May':5,'Jun':6,'Jul':7,'Aug':8,'Sep':9,'Oct':10,'Nov':11,'Dec':12}
def dateparse(d):
    if "to" in d:
        d_work = d.split(" to ")
        if '?' in d_work[1]:
            d_work = d_work[0]
        else:
            d_work = d_work[1]
    else:
        d_work = d
    d_work = d_work.replace(',','').split(' ')
    Y = int(d_work[-1])
    if len(d_work)==1:
        M = 1
        D = 1
        Y = int(d_work[0])
    elif len(d_work)==2:
        M = months[d_work[0]]
        D = 1
        Y = int(d_work[1])
    else:
        M = months[d_work[0]]
        try:
            D = int(d_work[1])
        except:
            D = 1
        Y = int(d_work[2])
    if Y<1970:
        return 0
    return float(datetime.datetime(Y,M,D).timestamp())

relevant['Aired'] = np.vectorize(dateparse,otypes=[float])(relevant['Aired'])
```

```python
def numconv(t):
    try:
        return float(t.split(" ")[0])
    except:
        return np.nan

relevant['Duration'] = np.vectorize(numconv,otypes=[float])(relevant['Duration'])
relevant['Episodes'] = np.vectorize(numconv,otypes=[float])(relevant['Episodes'])
```
